```python
import pandas as pd
import numpy as np
```

```python
df_sub = pd.read_csv('Rule_based_Confirmation.csv')
```

```python
def recreate_sub(group):
    print(group)
    group = group[group['score'] != 0]
    if(group.shape[0] == 0):
        return ""
    else:
        print(group)
        group = group.sort_values(by=['score'], ascending=False)
        print(group)
        return " ".join(group['hotel_id'])
```

```python
df_sub = df_sub.groupby('session_id').apply(lambda x: recreate_sub(x)).reset_index(name='item recommendations')
df_sub.head()
```
