```python
import csv

all_losses = []

for i in range(5):
    loss = 1.0 * i
    all_losses.append(loss)
    
with open('try_print_loss.csv', mode='w') as loss_file:
    file_writer = csv.writer(loss_file)
    file_writer.writerow(['#Epochs'])
    for loss in all_losses:
        print(str(loss))
        file_writer.writerow([loss])
```
