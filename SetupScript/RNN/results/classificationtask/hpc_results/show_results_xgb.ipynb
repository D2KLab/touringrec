```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plotn(n, perc):
    df_loss = pd.read_csv('./#' + str(n) + '/rnn_xg_' + str(perc) + '%_sub_loss.csv')
    df_acc = pd.read_csv('./#' + str(n) + '/rnn_xg_' + str(perc) + '%_sub_acc.csv')
    plt.figure()
    plt.plot(df_loss)
    plt.figure()
    plt.plot(df_acc)
```

```python
plotn(1, 10)
```
