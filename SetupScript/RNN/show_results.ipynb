```python
def plotn(n, perc):
    df_loss = pd.read_csv('./results/#' + str(n) + '/rnn_' + str(perc) + '%_sub_loss.csv')
    df_acc = pd.read_csv('./results/#' + str(n) + '/rnn_' + str(perc) + '%_sub_acc.csv')
    plt.figure()
    plt.plot(df_loss)
    plt.figure()
    plt.plot(df_acc)
```

```python
plotn(1, 1)
```
